```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
i = np.array([1, 0])
j = np.array([0, 1])

print(f"Producto punto entre i y j: {i @ j}")
```

```python
norma_i = np.linalg.norm(i)
norma_j = np.linalg.norm(j)
print(f"Norma de i: {norma_i}")
print(f"Norma de j: {norma_j}")
```

```python
v = np.array([5, 3])

v_reconstruido = 5 * i + 3 * j
print(f"¿La base estándar puede constuir este vector v? -> {np.array_equal(v, v_reconstruido)}")
```

```python
v1 = np.array([4, -2])
v2 = np.array([2, 4])

v3 = np.array([3, 3])

producto_punto_v1_v2 = v1 @ v2
print(f"Producto punto entre v1 y v2: {producto_punto_v1_v2}")

producto_punto_v1_v3 = v1 @ v3
print(f"Producto punto entre v1 y v3: {producto_punto_v1_v3}")
```

```python
norma_v1 = np.linalg.norm(v1)
norma_v2 = np.linalg.norm(v2)
print(f"Norma v1: {norma_v1}")
print(f"Norma v2: {norma_v2}")
```

```python
v1_normalizado = v1 / norma_v1
v2_normalizado = v2 / norma_v2
v3_normalizado = v3 / np.linalg.norm(v3)
print(f"Vector v1 normalizado: {v1_normalizado}")
print(f"Vector v2 normalizado: {v2_normalizado}")
```

```python
producto_punto_v1_v2_normalizado = v1_normalizado @ v2_normalizado
norma_v1_normalizado = np.linalg.norm(v1_normalizado)
norma_v2_normalizado = np.linalg.norm(v2_normalizado)
```

```python
print(f"Producto punto entre v1 y v2 normalizado: {producto_punto_v1_v2_normalizado}")
print(f"Norma v1 normalizado: {norma_v1_normalizado:.2f}")
print(f"Norma v2 normalizado: {norma_v2_normalizado:.2f}")
```

```python
print(f'Ángulo en radianes entre v2 y v3 (grados): {(np.arccos(v2_normalizado @ v3_normalizado)):.2f}')
```

```python
print(f'Ángulo entre v2 y v3 (grados): {np.degrees(np.arccos(v2_normalizado @ v3_normalizado)):.2f}')
```
