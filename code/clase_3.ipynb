### Matriz de Películas de Netflix

### Tensor de Píxeles de Imágenes RGB

```python
Tensor_pixeles_imagen=np.array([
    [[255, 0, 0], [0, 255, 0], [0, 0, 255]],
    [[255, 255, 0], [0, 255, 255], [255, 0, 255]],
    [[192, 192, 192], [128, 128, 128], [0, 0, 0]]
])
print(f'Tensor de Píxeles de Imagen RGB:\n{Tensor_pixeles_imagen}')
print(f'Dimensión del Tensor de Píxeles de Imagen RGB: {Tensor_pixeles_imagen.shape}')
print(f'Tipo de dato del Tensor de Píxeles de Imagen RGB: {Tensor_pixeles_imagen.dtype}')
print(f'Número de elementos del Tensor de Píxeles de Imagen RGB: {Tensor_pixeles_imagen.size}')
print(f'Tipo del objeto Tensor de Píxeles de Imagen RGB: {type(Tensor_pixeles_imagen)}')
print(f'Número de ejes o dimensiones del Tensor de Píxeles de Imagen RGB: {Tensor_pixeles_imagen.ndim}')
```

#### Visualización de Imágenes RGB

```python
plt.imshow(Tensor_pixeles_imagen)
plt.title('Visualización de Imagen RGB desde Tensor de Píxeles')
plt.axis('off')
plt.show()
```
