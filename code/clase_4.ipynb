```python
import numpy as np
import matplotlib.pyplot as plt
```

```python

calificaciones_reales=np.array([5,1])
calificaciones_predichas=np.array([4.5,2.5])
print(f'Calificaciones Reales: {calificaciones_reales}')
print(f'Calificaciones Predichas: {calificaciones_predichas}')

error_calculado=calificaciones_reales - calificaciones_predichas
print(f'Error Calculado: {error_calculado}')
```

```python
porcentaje_error=(error_calculado / calificaciones_reales) * 100
print(f'Porcentaje de Error: {porcentaje_error}')
```
