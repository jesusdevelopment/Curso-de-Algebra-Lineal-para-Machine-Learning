### Producto punto entre vectores

```python
u=np.array([1,2,3]) 
v=np.array([4,5,6])

producto_punto1=np.dot(u,v)
print(f'El producto punto entre u y v es: {producto_punto1}')

producto_punto2=u.dot(v)
print(f'El producto punto entre u y v (método 2) es: {producto_punto2}')

producto_punto_alternativo=u @ v
print(f'El producto punto entre u y v (alternativo) es: {producto_punto_alternativo}')

producto_interno=np.inner(u,v)
print(f'El producto interno entre u y v es: {producto_interno}')
```

```python
import gensim.downloader as api

print("Cargando el modelo preentrenado GloVe...")
word_vectors = api.load("glove-wiki-gigaword-50")
print("Modelo cargado exitosamente.")
```

```python
rey = word_vectors['king']
print(f'Vector de "king": {rey}')
print(f'Dimensión del vector de "king": {rey.shape}')
hombre = word_vectors['man']
reina = word_vectors['queen']
gato = word_vectors['cat']

similitud_coseno = lambda a, b: np.dot(a, b) / (np.linalg.norm(a) * np.linalg.norm(b))

sim_rey_hombre = similitud_coseno(rey, hombre)
sim_rey_reina = similitud_coseno(rey, reina)
sim_rey_gato = similitud_coseno(rey, gato)

print(f"similitud rey-hombre: {sim_rey_hombre:.2f}")
print(f"similitud rey-reina: {sim_rey_reina:.2f}")
print(f"similitud rey-gato: {sim_rey_gato:.2f}")
```
